```python
def get_data(path_file, sep="\t"):
    return pd.read_csv(path_file, sep=sep)
```

```python
def getCSV(resultat, path_output):
  """
  input : dict(result), str(path to csv)
  output : obj(df)
  """
  df = pd.DataFrame.from_dict(resultat, orient="index")
  # print(df)
  return df.to_csv(path_output, sep="\t", encoding="utf8")
```

```python
def getSum(contextes):
  sum = 0
  for i in contextes.values():
    sum += i
  return sum
```

```python
def getMax(contextes):
  nb_max = 0
  contexte_max = list()
  for contexte, nb in contextes.items():
    if nb > nb_max:
      nb_max = nb
      contexte_max = list()
      contexte_max.append(contexte)
    elif nb == nb_max:
      contexte_max.append(contexte)
  return contexte_max, nb_max
```

```python
def getScoreLocuteur_aAEI(path_to_dir='/content/drive/Shareddrives/ReseauNeurones/resume_resultats/non_troncons/aA/locuteurs/*', output='/content/drive/Shareddrives/ReseauNeurones/resume_resultats/resultat_contexte-'):
  resultat = dict()
  fileName_output = path_to_dir.split('/')[-4] + '_' + path_to_dir.split('/')[-3]
  output += fileName_output + ".csv"

  for path in glb.glob(path_to_dir):
    print(path)
    if path.endswith('.csv'):
      locuteur = path.split('/')[-1][:-4]
      # print(locuteur)
      data = get_data(path)
      contextes = dict()
      contextes_faux_a = dict()
      contextes_faux_A = dict()
      contextes_faux_E = dict()
      contextes_faux_I = dict()
      nb_bon_a, nb_bon_A = 0, 0
      nb_bon_E, nb_bon_I = 0, 0
      
      for index, row in data.iterrows():
          if row['son'] == "a":
            if row['proba_oral'] < row['proba_nasal']:
              if row['contextes'] in contextes_faux_a:
                contextes_faux_a[row['contextes']] += 1
              else: contextes_faux_a[row['contextes']] = 1
            else: nb_bon_a += 1

          elif row['son'] == "E":
            if row['proba_oral'] < row['proba_nasal']:
              if row['contextes'] in contextes_faux_E:
                contextes_faux_E[row['contextes']] += 1
              else: contextes_faux_E[row['contextes']] = 1
            else: nb_bon_E += 1

          elif row['son'] == "A":
            if row['proba_nasal'] < row['proba_oral']:
              if row['contextes'] in contextes_faux_A:
                contextes_faux_A[row['contextes']] += 1
              else: contextes_faux_A[row['contextes']] = 1
            else: nb_bon_A += 1

          else:
            if row['proba_nasal'] < row['proba_oral']:
              if row['contextes'] in contextes_faux_I:
                contextes_faux_I[row['contextes']] += 1
              else: contextes_faux_I[row['contextes']] = 1
            else: nb_bon_I += 1
      
      contexte_max_a, nb_max_a = getMax(contextes_faux_a)
      contexte_max_A, nb_max_A = getMax(contextes_faux_A)
      contexte_max_E, nb_max_E = getMax(contextes_faux_E)
      contexte_max_I, nb_max_I = getMax(contextes_faux_I)

      contextes['nb_faux_a'] = getSum(contextes_faux_a)
      contextes['contextes_a'] = "/".join(contextes_faux_a.keys())
      contextes['nb_bon_a'] = nb_bon_a
      contextes['nb_faux_A'] = getSum(contextes_faux_A)
      contextes['contextes_A'] = "/".join(contextes_faux_A.keys())
      contextes['nb_bon_A'] = nb_bon_A
      contextes['nb_faux_E'] = getSum(contextes_faux_E)
      contextes['contextes_E'] = "/".join(contextes_faux_E.keys())
      contextes['nb_bon_E'] = nb_bon_E
      contextes['nb_faux_I'] = getSum(contextes_faux_I)
      contextes['contextes_I'] = "/".join(contextes_faux_I.keys())
      contextes['nb_bon_I'] = nb_bon_I

      contextes['nb_faux_a_max'] = nb_max_a
      contextes['contexte_a_avec_fauxMax'] = "/".join(contexte_max_a)
      contextes['nb_faux_A_max'] = nb_max_A
      contextes['contexte_A_avec_fauxMax'] = "/".join(contexte_max_A)
      contextes['nb_faux_E_max'] = nb_max_E
      contextes['contexte_E_avec_fauxMax'] = "/".join(contexte_max_E)
      contextes['nb_faux_I_max'] = nb_max_I
      contextes['contexte_I_avec_fauxMax'] = "/".join(contexte_max_I)

    
      resultat[locuteur] = contextes

  return getCSV(resultat, output)
```

```python
def getMean(list):
  total = 0
  for i in list:
    total += i
  return total/len(list)
```

```python
def durationMean(data, segs):
  durees = {}
  if segs == "aA":
    duree_a, duree_A = list(), list()
    for index, row in data.iterrows():
      duree = row["temps_fin"] - row["temps_debut"]
      if row["son"] == "a":
        duree_a.append(duree)
      elif row["son"] == "A":
        duree_A.append(duree)
    durees["mean_a"] = getMean(duree_a)
    durees["mean_A"] = getMean(duree_A)

  elif segs == "aAEI":
    duree_a, duree_A = list(), list()
    duree_E, duree_I = list(), list()
    for index, row in data.iterrows():
      duree = row["temps_fin"] - row["temps_debut"]
      if row["son"] == "a":
        duree_a.append(duree)
      elif row["son"] == "A":
        duree_A.append(duree)
      elif row["son"] == "E":
        duree_E.append(duree)
      elif row["son"] == "I":
        duree_I.append(duree)
    durees["mean_a"] = getMean(duree_a)
    durees["mean_A"] = getMean(duree_A)
    durees["mean_E"] = getMean(duree_E)
    durees["mean_I"] = getMean(duree_I)

  return durees
```

```python
def getMeanF0(list):
  sum = 0
  for i in list:
    sum += i
  return sum / len(list)
```

```python
def f0Mean(data, segs, column):
  values = {}
  if segs == "aA":
    val_a, val_A = [], []
    for index, row in data.iterrows():
      if row[column] == "--undefined--": row[column] = 0
      else: row[column] = float(row[column])
      if row["son_x"] == "a" and row["son_y"] == row["son_x"]:
        val_a.append(row[column])
      elif row["son_x"] == "A" and row["son_y"] == row["son_x"]:
        val_A.append(row[column])
    values[column+"_a"] = getMeanF0(val_a)
    values[column+"_A"] = getMeanF0(val_A)

  elif segs == "aAEI":
    val_a, val_A = list(), list()
    val_E, val_I = list(), list()
    for index, row in data.iterrows():
      if row[column] == "--undefined--": row[column] = 0
      else: row[column] = float(row[column])
      if row["son_x"] == "a" and row["son_y"] == row["son_x"]:
        val_a.append(row[column])
      elif row["son_x"] == "A" and row["son_y"] == row["son_x"]:
        val_A.append(row[column])
      elif row["son_x"] == "E" and row["son_y"] == row["son_x"]:
        val_E.append(row[column])
      elif row["son_x"] == "I" and row["son_y"] == row["son_x"]:
        val_I.append(row[column])
    values[column+"_a"] = getMean(val_a)
    values[column+"_A"] = getMean(val_A)
    values[column+"_E"] = getMean(val_E)
    values[column+"_I"] = getMean(val_I)

  return values
```

```python
def tendance_f0(bas, eleve):
  if bas == 0 and eleve == 0: 
    tendance = "null"
  elif bas > eleve: 
    tendance = "bas"
  elif bas == eleve and bas != 0: 
    tendance = "egal"
  else: tendance = "eleve"
  return tendance
```
